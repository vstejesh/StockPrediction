```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.optimizers import Adam
```

```python
# from sklearn.preprocessing import MinMaxScaler
# train_close= train_df['close']
# test_close=test_df['close']

# # scaler=MinMaxScaler(feature_range=(0,1))

# # train_close=scaler.fit_transform(np.array(train_close).reshape(-1,1))
# # test_close=scaler.fit_transform(np.array(test_close).reshape(-1,1))


# scaler = MinMaxScaler()
# combined_data = np.concatenate((train_close, test_close), axis=0)
# scaled_combined_data = scaler.fit_transform(combined_data)
# train_close = scaled_combined_data[:len(train_close)]
# test_close = scaled_combined_data[len(train_close):]

from sklearn.preprocessing import MinMaxScaler
train_close= train_df['close']
test_close=test_df['close']

# scaler=MinMaxScaler(feature_range=(0,1))

# train_close=scaler.fit_transform(np.array(train_close).reshape(-1,1))
# test_close=scaler.fit_transform(np.array(test_close).reshape(-1,1))


scaler = MinMaxScaler()
# Convert the Series to NumPy arrays and reshape to 2D
combined_data = np.concatenate((train_close.values.reshape(-1, 1), test_close.values.reshape(-1, 1)), axis=0)
scaled_combined_data = scaler.fit_transform(combined_data)
train_close = scaled_combined_data[:len(train_close)]
test_close = scaled_combined_data[len(train_close):]
```

```python
train_close.shape
```

```python
time_step=350
X,y=[],[]
for i in range(time_step,len(train_close)):
    a = train_close[i-time_step:i, 0]
    # Append the sequence to dataX
    X.append(a)
    y.append(train_close[i, 0])
X=np.array(X)
y=np.array(y)
```

```python
X.shape
```

```python
X=X.reshape(X.shape[0],X.shape[1] ,1)
```

```python
model = Sequential()
model.add(LSTM(128,return_sequences='True', activation='relu', input_shape=(50, 1)))
model.add(LSTM(64,activation='relu'))
model.add(Dense(1))
optimizer = Adam(learning_rate=0.0001,clipvalue=1.0)
model.compile(optimizer=optimizer, loss='mean_squared_error')
```

```python
# X_test,y_test=[],[]
# for i in range(216):
#     a = scaled_combined_data[-i - time_step:i, 0]
#     X_test.append(a)
#     y_test.append(test_close[i, 0])

# X_test = np.array(X_test)
# y_test = np.array(y_test)

# X_test,y_test=[],[]
# for i in range(216):
#        a = scaled_combined_data[-566+i:215+i,0]
#        print(a.shape) # Print the shape of 'a'
#        X_test.append(a)
#        y_test.append(test_close[i, 0])

# X_test = np.array(X_test)
# y_test = np.array(y_test)
# print(X_test.shape) # Print the shape of X_test before reshaping
# X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
# print(X_test.shape) # Print the shape of X_test after reshaping

import numpy as np

# Assuming train_close and test_close are already scaled and defined
time_step = 50

# Reshape train_close and test_close to ensure they're 2D arrays
train_close = train_close.reshape(-1, 1)
test_close = test_close.reshape(-1, 1)

# Extract the last 350 values from train_close
last_train_window = train_close[-time_step:]

# Concatenate the last part of train_close with test_close
combined_test_data = np.concatenate((last_train_window, test_close), axis=0)

# Prepare X_test and y_test
X_test = []
y_test = []

# Generate sequences using combined_test_data
for i in range(time_step, len(combined_test_data)):
    X_test.append(combined_test_data[i-time_step:i, 0])  # Use the previous 350 values
    y_test.append(combined_test_data[i, 0])  # Predict the next value

# Convert lists to NumPy arrays
X_test = np.array(X_test)
y_test = np.array(y_test)

# Reshape X_test to fit the LSTM input format
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Print the shapes to verify
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
model.fit(X, y, epochs=25, batch_size=32,validation_data=(X_test,y_test))
```
